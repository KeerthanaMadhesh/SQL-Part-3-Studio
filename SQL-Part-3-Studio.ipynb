### **Event 1:** Women's History Month

Goal: Highlight Popular Womans Authors based on ratings from the Good Reads DB.

### **Write a query that returns tag\_id, the number of times each tag\_id is used and the tag\_name.**  

### Use the GROUP BY and HAVING clause to narrow your focus.

Try multiple keywords, ie woman vs female (female ended up giving better results)

```sql
-- Solution
SELECT t.tag_id, count(*) as usage_count, t.tag_name from dbo.tags t
join book_tags u on t.tag_id = u.tag_id
where lower(t.tag_name) like '%female%' or lower(t.tag_name) like '%woman%'
group by t.tag_id,t.tag_name
order by usage_count desc;
```

### **Which keyword(s) and tag(s) would you like to explore more?**

We need to focus on the keyword Women and related tags

### **Write a query to return authors, titles, ratings and** **selected tag(s)\_id.**

```sql
select b.authors, b.title, r.rating, t.tag_id from books b
join ratings r on b.book_id = r.book_id
join book_tags bt on b.book_id = bt.goodreads_book_id
join tags t on bt.tag_id = t.tag_id
where lower(t.tag_name) like '%women%'
order by r.rating desc;
```

```sql
-- Event 1 Query
select b.title, b.authors, b.best_book_id from books b 
join book_tags bt on b.best_book_id = bt.goodreads_book_id
where b.best_book_id = 1
order by b.best_book_id, b.title;
 
```

### **Summary of event 1:**

Showcased the collection of best selling boks by using query that joins the books and book\_tags tables

```sql
-- Event 2 Query
select b.title, b.authors, avg(r.rating) as avg_rating
from books b 
join ratings r on b.book_id = r.book_id
group by b.title, b.authors
having avg(r.rating) >= 4
order by avg_rating desc;
```
